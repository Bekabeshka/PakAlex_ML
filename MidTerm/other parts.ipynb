```python
import pandas as pd
import numpy as np
data = pd.read_csv("data.csv")
data.fillna(data.mean())
```

```python
from sklearn import linear_model
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression

scores = pd.read_csv("scores.csv")
scores = pd.read_csv("scores.csv", sep = ',\s*', header = None, engine = 'python')

fst_col = scores.iloc[:, 0]

X = data[['age', 'studytime', 'health']]
Y = fst_col

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)

regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

print('Liner Regression R squared: %.4f' % regressor.score(X_test, y_test))
print(regressor.intercept_, regressor.coef_)
```

```python
att = pd.read_csv('attendance.csv')
ss = pd.read_csv('school_support.txt')

data['attendance'] = att
data['school_support'] = ss
```

```python
wrong_tt = ((data['traveltime'] > 4) | (data['traveltime'] < 1))
data.loc[wrong_tt, 'traveltime'] = np.int_((data.loc[wrong_tt, 'traveltime'] % 4) + 1)
```

```python
wrong_age = (data['age'] > 1900)
data.loc[wrong_age, 'age'] = np.int_(2018 - data.loc[wrong_age, 'age'])
```

```python
def divide(arr): 
    pu = []
    td = []
    for i in arr:
        pu.append(np.int_(i / 10))
        td.append(i % 10)
    return pu, td

pu, td = divide(data['plans_universitypast_failures'])
data['plans_university'] = pu
data['past_failures'] = td
data = data.drop(columns='plans_universitypast_failures')
data
```
